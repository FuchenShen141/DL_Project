```python
from google.colab.patches import cv2_imshow

import cv2
import os
img1_root = '/content/drive/MyDrive/DR_imgae/image_filtering/other_test/NAF_NET'
img2_root = '/content/drive/MyDrive/DR_imgae/image_filtering/other_test/zero_dce'
out_root = '/content/drive/MyDrive/DR_imgae/image_filtering/ensemble_result/ZeroDCE_NAF'


for seq in range(1,6):
  file = ('out_%s.png' % (seq) )
  filename = os.path.join( img1_root, file )   
  image1=cv2.imread(filename)

  file = ('out_%s.png' % (seq) )
  filename = os.path.join( img2_root, file )   
  image2=cv2.imread(filename)

  image3=cv2.addWeighted(image1,0.7,image2,0.3,0)
  cv2_imshow(image3)

  file =('out_%s.png' % (seq))
  filename = os.path.join(out_root, file)
  cv2.imwrite(filename,image3)
```
